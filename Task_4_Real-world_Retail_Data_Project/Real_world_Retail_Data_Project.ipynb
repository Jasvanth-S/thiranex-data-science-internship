### Handle Missing Values

Based on the analysis of missing values, we will decide on an appropriate strategy for handling them. Common strategies include:
- **Imputation:** Filling missing values with a calculated statistic (mean, median, mode) or a specific value.
- **Deletion:** Removing rows or columns with a high percentage of missing values.

For numerical columns, we might impute with the mean or median. For categorical columns, we might use the mode or a placeholder like 'Unknown'.

```python
if df is not None:
    # Example: Impute numerical columns with median and categorical with mode/placeholder
    # Identify numerical and categorical columns that have missing values
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype in ['int64', 'float64']:
                # Impute numerical columns with the median
                median_val = df[col].median()
                df[col].fillna(median_val, inplace=True)
                print(f"Missing values in numerical column '{col}' imputed with median: {median_val}")
            else:
                # Impute categorical columns with the mode (most frequent value) or 'Unknown'
                # Using mode for demonstration; 'Unknown' can be used if appropriate
                mode_val = df[col].mode()[0]
                df[col].fillna(mode_val, inplace=True)
                # df[col].fillna('Unknown', inplace=True)
                print(f"Missing values in categorical column '{col}' imputed with mode: {mode_val}")

    print("\nAfter handling missing values:")
    print(df.isnull().sum().sum(), "total missing values remaining.")
else:
    print("Cannot handle missing values: DataFrame not loaded.")
```

### Detect and Handle Outliers

Outliers are data points that significantly differ from other observations. They can be legitimate extreme values or errors. Detecting and handling them is important as they can disproportionately affect statistical analyses and ML model training. Common methods include:
- **Visualization:** Box plots, scatter plots.
- **Statistical methods:** Z-score, IQR (Interquartile Range) method.

We will use the IQR method for a robust approach to identify and potentially cap/floor outliers in numerical columns.

```python
if df is not None:
    # Select numerical columns for outlier detection
    numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
    print(f"Numerical columns for outlier detection: {numerical_cols}")

    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
        if outliers_count > 0:
            print(f"Column '{col}': {outliers_count} outliers detected.")

            # Visualize outliers using a box plot
            plt.figure(figsize=(8, 4))
            sns.boxplot(x=df[col])
            plt.title(f'Box Plot of {col} (Outlier Detection)')
            plt.show()

            # Handling strategy: Capping (Winsorization) - replace outliers with bounds
            df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
            df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
            print(f"Outliers in '{col}' have been capped/floored to {lower_bound:.2f} and {upper_bound:.2f}.")
        else:
            print(f"Column '{col}': No significant outliers detected.")

    print("\nOutlier handling complete.")
else:
    print("Cannot detect/handle outliers: DataFrame not loaded.")
```

### Monthly Sales Trend

This line plot shows how sales evolve over months. It's excellent for spotting seasonality, growth trends, or significant drops in sales over time. For this, we'll assume a 'Date' or 'OrderDate' column exists and convert it to datetime.

```python
if df is not None and 'Sales' in df.columns:
    date_column = None
    # Try to find a common date column name
    for col in ['Date', 'OrderDate', 'TransactionDate', 'SaleDate']:
        if col in df.columns:
            date_column = col
            break

    if date_column:
        # Ensure the date column is in datetime format
        df[date_column] = pd.to_datetime(df[date_column], errors='coerce')
        # Drop rows where date conversion failed
        df.dropna(subset=[date_column], inplace=True)

        if not df.empty:
            df['Month'] = df[date_column].dt.to_period('M')
            monthly_sales = df.groupby('Month')['Sales'].sum()
            monthly_sales.index = monthly_sales.index.astype(str) # Convert PeriodIndex to string for plotting

            plt.figure(figsize=(14, 7))
            sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
            plt.title('Monthly Sales Trend', fontsize=16)
            plt.xlabel('Month', fontsize=12)
            plt.ylabel('Total Sales', fontsize=12)
            plt.xticks(rotation=45, ha='right')
            plt.grid(axis='both', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()

            print("\nWhat the chart shows: This line graph displays the aggregated sales performance month-over-month, allowing for identification of temporal patterns.")
            print("Important observations: We can identify seasonal peaks and troughs, overall sales growth or decline over the observed period, and any unusual spikes or drops that might correlate with specific events.")
            print("Business insight: Understanding monthly trends is crucial for demand forecasting, optimizing staffing levels, planning promotional activities, and managing inventory to avoid overstocking or stockouts.")
        else:
            print("No valid date data after cleaning for Monthly Sales Trend.")
    else:
        print("Error: No suitable date column ('Date', 'OrderDate', 'TransactionDate', 'SaleDate') found for Monthly Sales Trend. Please ensure your dataset has a date column.")
elif df is not None:
    print("Error: 'Sales' column not found for Monthly Sales Trend. Please check column names.")
else:
    print("Cannot plot Monthly Sales Trend: DataFrame not loaded.")
```

### Create Profit Margin

If we have 'Sales' and 'Cost' (or similar cost-related column) data, we can calculate `Profit Margin`. This is a direct measure of profitability and can be a strong predictor or a valuable insight itself.

```python
if df is not None:
    # Assuming 'Sales' and 'Cost' columns exist
    if 'Sales' in df.columns and 'Cost' in df.columns:
        df['Profit'] = df['Sales'] - df['Cost']
        # Calculate Profit Margin as (Profit / Sales) * 100, handle division by zero
        df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
        df['Profit_Margin'].fillna(0, inplace=True) # Fill NaN if Sales was 0
        print("Created 'Profit' and 'Profit_Margin' features.")
        display(df[['Sales', 'Cost', 'Profit', 'Profit_Margin']].head())
    elif 'Sales' in df.columns and 'Profit' in df.columns:
        # If 'Profit' already exists (e.g., from initial dataset), calculate Profit_Margin
        df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
        df['Profit_Margin'].fillna(0, inplace=True) # Fill NaN if Sales was 0
        print("Created 'Profit_Margin' feature using existing 'Sales' and 'Profit' columns.")
        display(df[['Sales', 'Profit', 'Profit_Margin']].head())
    else:
        print("Cannot create 'Profit' or 'Profit_Margin' features: 'Sales' or 'Cost' (or 'Profit') columns not found. Please ensure these columns exist or create them if possible.")
else:
    print("Cannot create profit-related features: DataFrame not loaded.")
```

## 7. Feature Selection

Feature selection is the process of choosing a subset of relevant features (variables) to use in building a machine learning model. This step is important for several reasons:
- **Simplifies Models:** Reduces complexity, making models easier to interpret.
- **Reduces Training Time:** Fewer data points mean faster training.
- **Enhances Generalization:** Decreases overfitting by removing irrelevant or redundant features.
- **Mitigates Dimensionality Issues:** Helps manage the 'curse of dimensionality' in high-dimensional datasets.

### Choose Input Features (X) and Target Variable (y)

In a supervised machine learning task like sales prediction, we need to define:
- **Target Variable (y):** This is the variable we want to predict. In our case, it will be `Sales`.
- **Input Features (X):** These are the independent variables that our model will use to make predictions. We will select relevant numerical and categorical columns from our cleaned and engineered dataset.

```python
if df is not None:
    # Define the target variable (y)
    # We are predicting 'Sales'
    if 'Sales' in df.columns:
        y = df['Sales']
        print(f"Target variable (y) selected: 'Sales'\n")
    else:
        print("Error: 'Sales' column not found. Cannot define target variable. Please ensure your dataset has a 'Sales' column.")
        y = None # Set y to None if 'Sales' column is missing

    # Define the input features (X)
    # We need to select columns that are relevant and have appropriate data types.
    # Drop the target variable itself and any identifier columns or highly correlated features if necessary.
    if y is not None:
        # Identify potential features (exclude target, and non-predictive/identifier columns)
        # This is a generic example; you will need to adjust based on your specific dataset columns.
        # Common columns to exclude: 'ID', 'Name', original date columns if 'Month'/'Year' are created
        features_to_exclude = ['Sales', 'Product', 'Category', 'Region', 'Date', 'OrderDate', 'TransactionDate', 'SaleDate']

        # Dynamically build the list of columns to consider as features
        candidate_features = [col for col in df.columns if col not in features_to_exclude]

        # For simplicity, let's start with numerical features and any one-hot encoded categorical features.
        # We'll need to handle categorical features (e.g., using one-hot encoding) before model training.
        # For now, let's select numerical features.
        X = df[candidate_features].select_dtypes(include=np.number)

        print("Selected numerical input features (X) before handling categorical variables:")
        display(X.head())
        print(f"\nNumber of features: {X.shape[1]}")

        # Important: For categorical features, we will need to apply one-hot encoding.
        # This step is typically done before splitting the data, or within a Pipeline.
        # For now, we will identify them and prepare for one-hot encoding.
        categorical_features = df[candidate_features].select_dtypes(include='object').columns.tolist()
        if categorical_features:
            print(f"\nIdentified categorical features for encoding: {categorical_features}")
            # Example of preparing for one-hot encoding (actual encoding will be done later)
            # X = pd.get_dummies(df[candidate_features], columns=categorical_features, drop_first=True)
        else:
            print("\nNo categorical features identified for encoding in the current selection.")

        # Re-align X and y if needed (e.g., after dropping NaNs for date conversion)
        X, y = X.align(y, join='inner', axis=0)
        print("\nShape of X after alignment:", X.shape)
        print("Shape of y after alignment:", y.shape)

    else:
        print("Cannot define input features: Target variable 'Sales' not found.")
else:
    print("Cannot perform feature selection: DataFrame not loaded.")
```

### Explanation: Features vs. Target

- **Features (Independent Variables / Predictors):** These are the inputs to our model. They are the characteristics or attributes of the data that we believe will help us predict the target variable. In our retail sales project, features could include `Month`, `Year`, `Profit_Margin`, `Region`, `Category`, `Discount`, `Quantity`, etc. We will encode categorical features (like `Region`, `Category`) into numerical formats (e.g., using one-hot encoding) so that machine learning algorithms can process them.

- **Target (Dependent Variable / Response Variable):** This is the output or the outcome that our model aims to predict. It is the variable whose value is influenced by the features. In this project, `Sales` is our target variable, as we want to forecast future sales based on the given features.

## 8. Split Dataset

Before training any machine learning model, it's crucial to split our dataset into two main parts:
- **Training Data:** Used to train the model, allowing it to learn the patterns and relationships between features and the target variable.
- **Testing Data:** Used to evaluate the model's performance on unseen data, providing an unbiased estimate of its generalization ability.

### Use `train_test_split()`

We will use scikit-learn's `train_test_split` function to divide our `X` (features) and `y` (target) into training and testing sets. A common split ratio is 80% for training and 20% for testing.

```python
from sklearn.model_selection import train_test_split

if 'X' in locals() and 'y' in locals() and X is not None and y is not None:
    # Before splitting, we need to handle categorical features within X.
    # For simplicity, let's one-hot encode all object type columns in X.
    # It's better to do this *before* splitting or use a pipeline.

    # Identify categorical columns in X (which currently only contains numerical features)
    # We need to go back to the original df for categorical columns that were not excluded
    # and then align them with X (numerical features).

    # Let's rebuild X to include one-hot encoded categorical features.
    # This ensures consistency and proper encoding before splitting.

    # Re-define features_to_exclude to only exclude non-predictive identifiers or original date cols
    features_to_exclude_final = ['Sales', 'Product', 'Date', 'OrderDate', 'TransactionDate', 'SaleDate'] # Exclude target and raw date cols

    # Select all potential features (numerical and categorical) that are not excluded
    X_full_df = df[[col for col in df.columns if col not in features_to_exclude_final and col != y.name]].copy()

    # Identify categorical columns in this expanded X_full_df
    categorical_cols_for_ohe = X_full_df.select_dtypes(include='object').columns.tolist()

    if categorical_cols_for_ohe:
        print(f"Applying One-Hot Encoding to categorical features: {categorical_cols_for_ohe}")
        X_full_df = pd.get_dummies(X_full_df, columns=categorical_cols_for_ohe, drop_first=True) # drop_first to avoid multicollinearity

    # Align X_full_df and y again after potential changes and encoding
    X_full_df, y = X_full_df.align(y, join='inner', axis=0)

    # Now perform the split
    X_train, X_test, y_train, y_test = train_test_split(X_full_df, y, test_size=0.2, random_state=42)

    print("\nDataset split into training and testing sets.")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"y_test shape: {y_test.shape}")
else:
    print("Error: X or y not defined. Please ensure feature selection was performed correctly.")
```

### Explanation: Why We Split the Dataset

Splitting the dataset is essential for:
1.  **Preventing Overfitting:** A model that performs exceptionally well on training data but poorly on new, unseen data is said to be overfit. By testing on a separate dataset, we can identify and mitigate overfitting.
2.  **Unbiased Performance Evaluation:** The test set serves as a proxy for real-world unseen data. Evaluating the model on this set provides a more realistic and unbiased measure of its predictive performance.
3.  **Generalization:** We want our model to generalize well, meaning it should be able to make accurate predictions on data it has never seen before. The train-test split helps us assess this capability.

## 9. Build Machine Learning Models

In this section, we will train several machine learning models to predict retail sales. We will focus on regression models, as our target variable ('Sales') is continuous. We will use a Linear Regression model as a baseline, followed by more complex tree-based models like Decision Tree Regressor and Random Forest Regressor.

### Train Linear Regression Model

Linear Regression is a fundamental supervised learning algorithm that models the relationship between a dependent variable (sales) and one or more independent variables (features) by fitting a linear equation to the observed data. It serves as a good baseline due to its simplicity and interpretability.

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
if 'X_train' in locals() and 'y_train' in locals():
    lr_model = LinearRegression()

    # Train the model
    lr_model.fit(X_train, y_train)
    print("Linear Regression model trained successfully.")

    # Store models for later comparison
    models = {'Linear Regression': lr_model}
else:
    print("Error: Training data (X_train, y_train) not found. Please ensure the dataset was split correctly.")
    models = {}
```

## 10. Model Evaluation

Model evaluation is a critical step to understand how well our trained machine learning models generalize to new, unseen data. We will use several regression metrics and visualizations to assess their performance and compare them.

### Evaluate Models using MAE, MSE, RMSE, R² Score

We will calculate common regression evaluation metrics for each model:
-   **Mean Absolute Error (MAE):** The average of the absolute differences between predictions and actual values. It gives an idea of the magnitude of errors without considering their direction.
-   **Mean Squared Error (MSE):** The average of the squared differences between predictions and actual values. It penalizes larger errors more heavily than MAE.
-   **Root Mean Squared Error (RMSE):** The square root of MSE. It is in the same units as the target variable, making it more interpretable than MSE.
-   **R-squared (R² Score):** Represents the proportion of the variance in the dependent variable that is predictable from the independent variables. A higher R² indicates a better fit for the model.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

if 'models' in locals() and 'X_test' in locals() and 'y_test' in locals():
    results = {}
    for name, model in models.items():
        y_pred = model.predict(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)

        results[name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2 Score': r2}

        print(f"\n--- {name} Evaluation ---")
        print(f"MAE: {mae:.2f}")
        print(f"MSE: {mse:.2f}")
        print(f"RMSE: {rmse:.2f}")
        print(f"R2 Score: {r2:.2f}")

    # Convert results to a DataFrame for easy comparison
    results_df = pd.DataFrame(results).T.sort_values(by='R2 Score', ascending=False)
    print("\n--- Model Comparison ---")
    display(results_df)

    # Store predictions for plotting
    predictions = {}
    for name, model in models.items():
        predictions[name] = model.predict(X_test)

else:
    print("Error: Models or test data not found. Please ensure models were trained and data was split correctly.")
```

### Create Actual vs Predicted Plot

An Actual vs. Predicted plot visually compares the model's predictions against the true values. Ideally, points should fall close to a 45-degree line, indicating perfect predictions. Deviations from this line highlight prediction errors.

```python
if 'predictions' in locals() and 'y_test' in locals():
    plt.figure(figsize=(15, 7))
    for name, y_pred in predictions.items():
        plt.scatter(y_test, y_pred, alpha=0.5, label=f'{name} Predictions')

    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect Prediction Line')
    plt.title('Actual vs. Predicted Sales for Test Data', fontsize=16)
    plt.xlabel('Actual Sales', fontsize=12)
    plt.ylabel('Predicted Sales', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

    print("\nWhat the chart shows: This scatter plot compares the actual sales values (on the x-axis) with the predicted sales values (on the y-axis) for each of our models. The red dashed line represents where perfect predictions would lie.")
    print("Important observations: Points closer to the red line indicate more accurate predictions. A wider spread of points away from the line suggests higher error rates. We can also observe if a model consistently over- or under-predicts.")
    print("Business insight: This visualization helps in understanding the model's reliability and its ability to capture the true sales trends. It can highlight areas where the model struggles, such as predicting very high or very low sales values.")
else:
    print("Cannot create Actual vs Predicted Plot: Predictions or test data not found.")
```

### Feature Importance Plot (for Tree-based Models)

For tree-based models like Decision Tree and Random Forest, we can extract and visualize feature importance. This shows which features contributed most to the model's predictions, providing valuable insights into the drivers of sales.

```python
if 'models' in locals() and 'X_train' in locals():
    # Check if any tree-based model is present
    tree_models_found = False
    for name, model in models.items():
        if isinstance(model, (DecisionTreeRegressor, RandomForestRegressor)):
            tree_models_found = True
            print(f"\n--- Feature Importance for {name} ---")
            feature_importances = pd.Series(model.feature_importances_, index=X_train.columns)
            top_10_features = feature_importances.nlargest(10)

            if not top_10_features.empty:
                plt.figure(figsize=(12, 7))
                sns.barplot(x=top_10_features.values, y=top_10_features.index, palette='plasma')
                plt.title(f'Top 10 Feature Importances for {name}', fontsize=16)
                plt.xlabel('Importance', fontsize=12)
                plt.ylabel('Feature', fontsize=12)
                plt.tight_layout()
                plt.show()

                print("\nWhat the chart shows: This bar chart displays the relative importance of the top features identified by the tree-based model. Higher values indicate greater influence on the model's predictions.")
                print("Important observations: Features with higher importance scores are considered more impactful in determining sales outcomes. This can confirm hypotheses about sales drivers or reveal unexpected relationships.")
                print("Business insight: Understanding feature importance helps in prioritizing business efforts. For example, if 'Discount' is a highly important feature, it suggests that pricing strategies significantly affect sales. This can guide marketing, product development, and operational decisions.")
            else:
                print(f"No feature importances found for {name}. It might not have been fitted or has no features.")

    if not tree_models_found:
        print("No tree-based models (Decision Tree Regressor, Random Forest Regressor) found for feature importance plotting.")
else:
    print("Cannot create Feature Importance Plot: Models or training data not found.")
```
